```python
import pandas as pd
import matplotlib.pyplot as plt

# 讀取數據集並指定列的數據類型
data_path = r'E:\91APP_DataSet_2023\91APP_DataSet_2023\OrderData.csv'
dtype = {'OrderDateTime': str, 'Column2': int, 'Column3': float, 'Column4': str, 'Column5': str, 'Column6': str}
df = pd.read_csv(data_path, dtype=dtype)

# 將"OrderDateTime"字段轉換為日期時間類型
df['OrderDateTime'] = pd.to_datetime(df['OrderDateTime'])

# 提取年份
df['Year'] = df['OrderDateTime'].dt.year

# 統計每年的訂單量
order_counts = df['Year'].value_counts().sort_index()

# 繪制年份和訂單量關系圖
plt.figure(figsize=(8, 6))  # 調整圖表尺寸
plt.plot(order_counts.index, order_counts.values, marker='o')
plt.xlabel('Year')
plt.ylabel('Order Count')
plt.title('Order Count by Year')
plt.xticks(order_counts.index)  # 設置 x 軸刻度為年份
plt.xticks(rotation=45)  # 旋轉 x 軸刻度標簽
plt.show()
```
